# Lec 53 Fill in NaN values with fillna() method

```python
nba =  pd.read_csv("nba.csv")
```

```python
nba.head()
```

### try to use fillna() only on series rather than dataframe as a whole
### because you may not wanna replace NaN with a value that doesn't fit in Salary and College column at the same time

# Lec 54 Convert data type of a column using astype() method

# Lec 58 Rank Values with the .rank() Method
#### in order to make rank() method work, get rid of all null values first

```python
nba.dropna(how="all")
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba.head()
```

```python
nba["Salary"].rank(ascending=False).astype("int").head()
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending=False).astype("int")
```

```python
nba.head()
```
